### Criando nosso banco de dados

```python

from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma.vectorstores import Chroma
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(
  model = "mxbai-embed-large",
)

PASTA_PDFS = "base"
```

```python
def carregar_pdfs():
    pasta = PASTA_PDFS
    carregador = PyPDFDirectoryLoader(pasta, glob="**/*.pdf")
    documentos = carregador.load()
    return documentos
```

```python
def dividir_chunks(documentos):
    separador = RecursiveCharacterTextSplitter(
        # Tamanho do chunk em número de caracteres ("qual o tamanho maximo que minha para minha informação estar dentro?")
        chunk_size = 2000,
        chunk_overlap = 200,
        length_function = len,
        add_start_index = True,
    )
    chunks = separador.split_documents(documentos)
    # print(len(chunks))
    return chunks
```

```python
def vetorizar(chunks):
    db = Chroma.from_documents(chunks, embeddings, persist_directory="db")
    print("Banco de dados Criado")
```

```python
def criar_db():
    # carregar pdfs
    documentos = carregar_pdfs()
    # dividir documentos em chunks
    chunks = dividir_chunks(documentos)
    # vetorização e embedding
    vetorizar(chunks)
```

```python
criar_db()
```
